## Data cleaning

### Removing punctuation

```python
def remove_punctuation(text):
    '''a function for removing punctuation'''
    import string
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)
```

```python
# Apply remove_punctuation to each row of data['text'] 
data['text'] = data['text'].apply(remove_punctuation)
```

### Removing stopwords

```python
def remove_stopwords(text):
    from nltk.corpus import stopwords
    '''a function for removing the stopword'''
    sw = stopwords.words('english')
    # removing the stop words and lowercasing the selected words
    text = [word.lower() for word in text.split() if word.lower() not in sw]
    # joining the list of words with space separator
    return " ".join(text)
```

```python
# Apply remove_stopwords to each row of data['text'] 
data['text'] = data['text'].apply(remove_stopwords)
```

### Stemming words

```python
def stemming(text): 
    '''a function which stems each word in the given text'''
    from nltk.stem.snowball import SnowballStemmer
    stemmer = SnowballStemmer("english")
    text = [stemmer.stem(word) for word in text.split()]
    return " ".join(text) 
```

```python
# Apply remove_punctuation to each row of data['text'] 
data['text'] = data['text'].apply(stemming)
```

### Utility functions

```python
def pretty_print_cm(cm, labels, normalize=False):
    """pretty print confusion matrices"""
    import seaborn as sns
    from sklearn.preprocessing import scale
    
    title = 'Confusion matrix'
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        title = ' '.join(['Normalized', title])
    
    conf_mat = pd.DataFrame(cm, columns=labels, index=labels)
    conf_mat.index.name = 'True \ Predicted'
    ax = sns.heatmap(conf_mat, annot=True, cmap='Blues', fmt='g')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('True')
    ax.set_title(title)
    plt.show()
```

```python
def generate_prediction_results(y_true, y_pred):
    """generate classification report and confusion matrix on a given true and predicted values"""
    from sklearn.metrics import classification_report, confusion_matrix

    conf_mat = confusion_matrix(y_true, y_pred)
    class_report = classification_report(y_true, y_pred, output_dict=True)
    return conf_mat, class_report
```

## Recap and performance comparison

### Utility functions

```python
def metric_recap(classifiers_names, metric, *reports):
    '''
    returns a recap of a given metric of each classifier by average
    Input:  -list of strings that represents the classifiers names, 
            -metric name, 
            -the classification report of each classifier
    Output: pandas dataframe of the given metric
    '''
    micro_recap = []
    macro_recap = []
    weighted_recap = []

    for report in list(reports):
        micro_recap.append(report['micro avg'][metric])
        macro_recap.append(report['macro avg'][metric])
        weighted_recap.append(report['weighted avg'][metric])

    metric_recap_df = pd.DataFrame(
        [micro_recap, macro_recap, weighted_recap],
        columns=classifiers_names,
        index=['micro avg', 'macro avg', 'weighted avg'])
    metric_recap_df = metric_recap_df.T
    metric_recap_df.index.name = metric

    return metric_recap_df
```

```python
def metric_plot(metric_recap_df):
    """
    function that plots the given metric recap dataframe 
    """
    title = metric_recap_df.index.name
    plt.rcParams['figure.figsize'] = (24, 10)
    plt.rcParams['font.size'] = 20

    plt.plot(
        metric_recap_df['micro avg'],
        '-o',
        label='micro avg',
        linewidth=4,
        markersize=12)
    plt.plot(
        metric_recap_df['macro avg'],
        '-*',
        label='macro avg',
        linewidth=4,
        markersize=12)
    plt.plot(
        metric_recap_df['weighted avg'],
        '-+',
        label='weighted avg',
        linewidth=4,
        markersize=12)
    plt.xlabel = 'classifiers'
    plt.ylabel = 'metric avg'

    plt.legend(loc='lower right', prop={'size': 16})
    plt.title(title)
    plt.show()
```

### Performance comparison

```python
# Generating metrics' recap dataframes
classifiers_names = ['NB', 'SVM', 'RF', 'MLP']

precision_recap = metric_recap(classifiers_names, 'precision',
                         NB_classification_report, SVM_classification_report,
                         RF_classification_report, MLP_classification_report)
recall_recap = metric_recap(classifiers_names, 'recall',
                         NB_classification_report, SVM_classification_report,
                         RF_classification_report, MLP_classification_report)
f1score_recap = metric_recap(classifiers_names, 'f1-score',
                         NB_classification_report, SVM_classification_report,
                         RF_classification_report, MLP_classification_report)
```

```python
metric_plot(precision_recap)
metric_plot(recall_recap)
metric_plot(f1score_recap)
```
